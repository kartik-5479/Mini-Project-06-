* Ques 1. Create a file that contains 1000 lines of random strings.

```python
import random
import string

# Open a file to write
file = open("random_strings.txt", "w")

# Repeat 1000 times
for i in range(1000):
    # Make a random string of 10 characters
    s = ''.join(random.choices(string.ascii_letters, k=10))
    # Write the string to the file
    file.write(s + "\n")

# Close the file
file.close()

print("File created with 1000 random strings.")
```

* Ques 2. Create a file that contains multiple lines of random strings and file size must be 5 MB.

```python
import random
import string

file = open("random_5mb.txt", "w")

size_in_bytes = 5 * 1024 * 1024  # 5 MB
written_bytes = 0

while written_bytes < size_in_bytes:
    line = ''.join(random.choices(string.ascii_letters + string.digits, k=100))  # 100 characters
    file.write(line + "\n")
    written_bytes += len(line) + 1  # +1 for newline character

file.close()

print("File 'random_5mb.txt' created with size around 5 MB.")
```

* Ques 3. Create 10 files that contains multiple lines of random strings and file size of each file must be 5 MB.

```python
import os
import random
import string

def generate_random_string(length=100):
    return ''.join(random.choices(string.ascii_letters + string.digits, k=length))

def create_file(filename, target_size_mb=5):
    target_size = target_size_mb * 1024 * 1024  # Convert MB to bytes
    with open(filename, 'w') as f:
        current_size = 0
        while current_size < target_size:
            line = generate_random_string(100) + '\n'  # Each line ~101 bytes
            f.write(line)
            current_size += len(line)

# Create 10 files
for i in range(1, 11):
    filename = f"random_file_{i}.txt"
    create_file(filename)
    print(f"Created {filename} (5 MB)")
```

* Ques 4. Create 5 files of size 1GB, 2GB, 3GB, 4GB and 5GB; file contains multiple lines of random strings.

```python
import os
import random
import string

def generate_random_string(length=100):
    return ''.join(random.choices(string.ascii_letters + string.digits, k=length))

def create_large_file(filename, size_in_gb):
    target_size = size_in_gb * 1024 * 1024 * 1024  # Convert GB to bytes
    with open(filename, 'w') as f:
        current_size = 0
        while current_size < target_size:
            line = generate_random_string(100) + '\n'  # ~101 bytes per line
            f.write(line)
            current_size += len(line)

# File sizes in GB
file_sizes = [1, 2, 3, 4, 5]

# Create the files
for size in file_sizes:
    filename = f"large_file_{size}GB.txt"
    print(f"Creating {filename} ...")
    create_large_file(filename, size)
    print(f"Created {filename} ({size} GB)")
```

* Ques 5. Convert all the files of Q4 into upper case one by one.

```python
def convert_file_to_uppercase(filename):
    temp_filename = filename + ".tmp"

    with open(filename, 'r') as infile, open(temp_filename, 'w') as outfile:
        for line in infile:
            outfile.write(line.upper())

    # Replace original file with the uppercased version
    import os
    os.replace(temp_filename, filename)
    print(f"Converted {filename} to uppercase.")

# File names to convert
file_sizes = [1, 2, 3, 4, 5]
for size in file_sizes:
    filename = f"large_file_{size}GB.txt"
    convert_file_to_uppercase(filename)
```

* Ques 6. Convert all the files of Q5 into upper case parallel using multi-threading.

```python
import threading
import os

def convert_file_to_uppercase(filename):
    temp_filename = filename + ".tmp"

    with open(filename, 'r') as infile, open(temp_filename, 'w') as outfile:
        for line in infile:
            outfile.write(line.upper())

    os.replace(temp_filename, filename)
    print(f"[{filename}] conversion to uppercase done.")

# List of file names
file_sizes = [1, 2, 3, 4, 5]
file_names = [f"large_file_{size}GB.txt" for size in file_sizes]

# Create and start a thread for each file
threads = []
for file in file_names:
    t = threading.Thread(target=convert_file_to_uppercase, args=(file,))
    threads.append(t)
    t.start()

# Wait for all threads to finish
for t in threads:
    t.join()

print("All files converted to uppercase in parallel.")
```

* Ques 7. Create a random dataset of 500 rows and 10 columns.
Columns 1 to 4 are defined between [-10, 10];
Columns 5 to 8 are defined between [10, 20];
Columns 9 to 10 are defined between [-100, 100].
Apply following clustering algorithms; determine the optimal number of clusters and plot distance metric graph using each algorithm.
(i) K-Mean clustering
(ii) Hierarchical clustering

```python
pip install numpy pandas matplotlib seaborn scikit-learn scipy
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import pairwise_distances_argmin_min

# Step 1: Create dataset
np.random.seed(42)  # for reproducibility

data = np.hstack([
    np.random.uniform(-10, 10, (500, 4)),     # Columns 1 to 4
    np.random.uniform(10, 20, (500, 4)),      # Columns 5 to 8
    np.random.uniform(-100, 100, (500, 2))    # Columns 9 to 10
])

df = pd.DataFrame(data, columns=[f'Feature_{i+1}' for i in range(10)])
print(df.head())

# Optional: Standardize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 2: K-Means Clustering with Elbow Method
inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia (Within-cluster Sum of Squares)')
plt.title('K-Means Elbow Method')
plt.grid(True)
plt.show()

# Step 3: Hierarchical Clustering with Dendrogram
linked = linkage(scaled_data, method='ward')

plt.figure(figsize=(12, 6))
dendrogram(linked, truncate_mode='lastp', p=30, leaf_rotation=90, leaf_font_size=12, show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Cluster Size')
plt.ylabel('Distance')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create random dataset
np.random.seed(42)
data = np.random.uniform(-100, 100, size=(600, 15))
df = pd.DataFrame(data, columns=[f'Col_{i+1}' for i in range(15)])

# Step 2 (i): Scatter plot between Column 5 and Column 6
plt.figure(figsize=(8, 5))
plt.scatter(df['Col_5'], df['Col_6'], alpha=0.6, color='blue')
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.title('Scatter Plot: Column 5 vs Column 6')
plt.grid(True)
plt.show()

# Step 3 (ii): Histogram of all columns in a single graph
df.plot(kind='hist', bins=30, figsize=(10, 6), alpha=0.5, legend=True)
plt.title('Histogram of All Columns')
plt.xlabel('Value')
plt.grid(True)
plt.show()

# Step 4 (iii): Box plot of all columns in a single graph
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, orient='h', palette='Set3')
plt.title('Boxplot of All Columns')
plt.xlabel('Value')
plt.grid(True)
plt.tight_layout()
plt.show()
```

* # Ques 9. Create a random dataset of 500 rows and 5 columns:
All the values are defined between [5,10].
Perform the following operations:
####(i) Perform t-Test on each column.
####(ii) Perform Wilcoxon Signed Rank Test on each column.
####(iii) Perform Two Sample t-Test and Wilcoxon Rank Sum Test on Column 3 and Column 4

```python
pip install numpy pandas scipy
```

```python
import numpy as np
import pandas as pd
from scipy import stats

# Step 1: Create random dataset
np.random.seed(0)  # For reproducibility
data = np.random.uniform(5, 10, size=(500, 5))
df = pd.DataFrame(data, columns=['Col1', 'Col2', 'Col3', 'Col4', 'Col5'])

# Step 2: Perform t-test and Wilcoxon Signed Rank Test on each column
print("One-Sample t-Test and Wilcoxon Signed-Rank Test (vs. Mean = 7.5):")
for col in df.columns:
    t_stat, t_p = stats.ttest_1samp(df[col], 7.5)
    try:
        w_stat, w_p = stats.wilcoxon(df[col] - 7.5)
    except ValueError as e:
        w_stat, w_p = None, None  # Wilcoxon may fail if all differences are zero
    print(f"\n{col}:")
    print(f"  t-Test      t-statistic = {t_stat:.4f}, p-value = {t_p:.4f}")
    print(f"  Wilcoxon    w-statistic = {w_stat}, p-value = {w_p}")

# Step 3: Two-Sample t-Test and Wilcoxon Rank-Sum Test on Col3 and Col4
print("\nTwo-Sample t-Test and Wilcoxon Rank-Sum Test between Col3 and Col4:")
t2_stat, t2_p = stats.ttest_ind(df['Col3'], df['Col4'])
r_stat, r_p = stats.ranksums(df['Col3'], df['Col4'])

print(f"  Two-Sample t-Test        t-statistic = {t2_stat:.4f}, p-value = {t2_p:.4f}")
print(f"  Wilcoxon Rank-Sum Test   statistic = {r_stat:.4f}, p-value = {r_p:.4f}")
```
